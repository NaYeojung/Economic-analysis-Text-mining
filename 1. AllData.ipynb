```python
import pandas as pd
```

```python
df = pd.read_csv('crawling/mpb_list.csv')
df.head()
```

```python
df_2=df[['date','title']]
df_2['content']=df['Economic Situation']+df['Monetary Policy']
df_2
```

```python
import re

df_2['date'] = pd.to_datetime(df_2['date'])
df_filtered = df_2[df_2['date'] >= '2014-01-01']
for i in range(len(df_filtered['date'])):

    text=df_filtered.iloc[i,1]
    match = re.search(r'\(([^()]+)\)\s*$', text)

    if match:
        extracted_text = match.group(1)
    df_filtered.iloc[i,0]=extracted_text


df_filtered
```

```python
df_filtered=df_filtered[df_filtered['content']!='[][]']
# df_filtered.to_csv('minutes_test.csv',encoding='utf-8-sig')
```

```python
df2=pd.read_csv('crawling/financial_report.csv')
df2=df2.drop('company',axis=1)
df2=df2.drop('view',axis=1)
df2 = df2.sort_values(by='date')
df2.reset_index(drop=True, inplace=True)
df2['date'] = pd.to_datetime(df2['date'], format='%y.%m.%d')
df2['date'] = df2['date'].dt.strftime('%Y-%m-%d')
df2
```

```python
import re

def clean_text(text):
    pattern = (
        r"^\s*뉴스\s*[0-9]{4}년\s*[0-9]{1,2}월\s*[0-9]{1,2}일\s*.*?\n"
        r"|.*기자.*\n©\s*[0-9]{4}\s*(뉴스사|서울경제)\s*.*$"
    )
    text = re.sub(pattern, '', text, flags=re.MULTILINE | re.DOTALL)

    return text.strip()
```

```python
df3=pd.read_csv('crawling/news_list.csv')
df3 = df3.sort_values(by='date')
df3['date']= pd.to_datetime(df3['date'])
df3.reset_index(drop=True, inplace=True)

for i in range(len(df3['content'])):
    text=df3.iloc[i,2]
    cleaned_text=clean_text(text)
    df3.iloc[i,2]=cleaned_text

df3
```

```python
all_df=pd.concat([df_filtered,df2,df3])
all_df = all_df.sort_values(by='date')
all_df.reset_index(drop=True, inplace=True)
all_df
```

```python
all_df.to_csv('all_data_df.csv', encoding='utf-8-sig')
```
